```python
%load_ext autoreload
import numpy as np
import DP.dMDP as MDP
%autoreload 2
```

## Defining MDP parameters

```python
n_state = 3
n_action = 3
```

```python
dyn = np.zeros((n_state, n_state, n_action))
reward = np.zeros((n_state, n_action))
```

```python
gamma = 0.95
tol = 0.01
```

```python
for i in range(n_state):
    for j in range(n_state):
        for k in range(n_action):
            prob = input(f"Probability of transition from state {i} to state {j} from action a{k}: ")
            dyn[i,j,k] = prob
```

```python
for i in range(n_state):
        for k in range(n_action):
            rew = input(f"Reward of action a{k} from state {i}: ")
            reward[i,k] = rew
```

```python
reward = np.asarray([
       [ 0.  ,  0.  ,  0.05],
       [ 0.  ,  0.  ,  0.  ],
       [ 0.  ,  1.  ,  0.9 ]])
```

```python
dyn = np.asarray([
       [[ 0.55,  0.3 ,  1.  ],
        [ 0.45,  0.7 ,  0.  ],
        [ 0.  ,  0.  ,  0.  ]],

       [[ 1.  ,  0.  ,  0.  ],
        [ 0.  ,  0.4 ,  1.  ],
        [ 0.  ,  0.6 ,  0.  ]],

       [[ 0.  ,  0.  ,  0.  ],
        [ 1.  ,  0.6 ,  0.  ],
        [ 0.  ,  0.4 ,  1.  ]]])
```

## Value iteration

```python
model = MDP.dMDP(n_state, n_action, dyn, reward, gamma, tol, solver = 'VI')

value_init = np.zeros(n_state)
model.solve(value_init = value_init)
```

```python
model.plot_error()
```

## Policy iteration

```python
model = MDP.dMDP(n_state, n_action, dyn, reward, gamma, tol, solver = 'PI')

policy_init = np.zeros(n_state, dtype = int)
model.solve(policy_init = policy_init)
```

```python
model.plot_error()
```
