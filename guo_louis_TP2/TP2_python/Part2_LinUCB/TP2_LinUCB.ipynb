```python
%load_ext autoreload
import numpy as np
import arms
import UCB
import matplotlib.pyplot as plt
from linearmab_models import ToyLinearModel, ColdStartMovieLensModel
from tqdm import tqdm
from linMAB import linMAB
%autoreload 2
```

## 2. Linear Bandit on Real Data

```python
random_state = np.random.randint(0, 24532523)
# model = ToyLinearModel(
#     n_features=8,
#     n_actions=20,
#     random_state=random_state,
#     noise=0.1)

model = ColdStartMovieLensModel(
    random_state=random_state,
    noise=0.1
)
```

```python
n_a = model.n_actions
d = model.n_features

# Number of trials and simulations
T = 6000
nb_simu = 2 
lambd = 1
```

```python
alphas = np.arange(2, 25, 2)
alg_name = 'LinUCB'

mean_regret = dict.fromkeys(alphas, 0)
mean_norms = dict.fromkeys(alphas, 0)

for alpha in alphas:
    for k in tqdm(range(nb_simu), desc="Simulating {}".format(alg_name)):
        linUCB = linMAB(model, alpha = alpha, lambd = 1, T = T)
        regret, norm_dist = linUCB.run(mode = 'linUCB')
        mean_regret[alpha] += regret / nb_simu
        mean_norms[alpha] += norm_dist / nb_simu
```

```python
f, ax = plt.subplots(1, figsize = (15,8))
data_plots = [mean_norms[alpha] for alpha in alphas]
labels = [r"$\alpha = {}$".format(alpha) for alpha in alphas]
title = r'Metric: $\theta$ estimation'
ylabel = r'$d(\theta, \hat{\theta})$'
xlabel = 'Rounds'
linUCB.plot_standard(data_plots, labels, [0], title, ylabel, xlabel, ax)
plt.show()
```

```python
f, ax = plt.subplots(1, figsize = (15,8))
data_plots = [mean_regret[alpha].cumsum() for alpha in alphas]
labels = [r"$\alpha = {}$".format(alpha) for alpha in alphas]
title = r'Metric: Cumulative regret'
ylabel = r'Cumulative regret'
xlabel = 'Rounds'
linUCB.plot_standard(data_plots, labels, [0], title, ylabel, xlabel, ax)
plt.show()
```

```python
alg_name = 'linUCB'
nb_simu = 50
mean_regret_linUCB = np.zeros((T))
mean_norms_linUCB = np.zeros((T))

for k in tqdm(range(nb_simu), desc="Simulating {}".format(alg_name)):
    linUCB = linMAB(model, alpha = 2, lambd = 1, T = T)
    regret, norm_dist = linUCB.run(mode = alg_name)
    mean_regret_linUCB += regret / nb_simu
    mean_norms_linUCB += norm_dist / nb_simu
```

```python
alg_name = 'random'
nb_simu = 50
mean_regret_random = np.zeros((T))
mean_norms_random = np.zeros((T))

for k in tqdm(range(nb_simu), desc="Simulating {}".format(alg_name)):
    random_policy = linMAB(model, alpha = 2, lambd = 1, T = T)
    regret, norm_dist = random_policy.run(mode = alg_name)
    mean_regret_random += regret / nb_simu
    mean_norms_random += norm_dist / nb_simu
```

```python
alg_name = 'eps_greedy'
nb_simu = 50
mean_regret_greedy = np.zeros((T))
mean_norms_greedy = np.zeros((T))

for k in tqdm(range(nb_simu), desc="Simulating {}".format(alg_name)):
    eps_greedy = linMAB(model, epsilon = 0.1, lambd = 1, T = T)
    regret, norm_dist = eps_greedy.run(mode = alg_name)
    mean_regret_greedy += regret / nb_simu
    mean_norms_greedy += norm_dist / nb_simu
```

```python
f, ax = plt.subplots(1, figsize = (15,8))
data_plots = [mean_norms_linUCB, mean_norms_random, mean_norms_greedy]
labels = ['linUCB', 'random', 'greedy']
title = r'Metric: $\theta$ estimation'
ylabel = r'$d(\theta, \hat{\theta})$'
xlabel = 'Rounds'
linUCB.plot_standard(data_plots, labels, [0], title, ylabel, xlabel, ax)
plt.show()
```

```python
f, ax = plt.subplots(1, figsize = (15,8))
data_plots = [mean_regret_linUCB.cumsum(), mean_regret_random.cumsum(), mean_regret_greedy.cumsum()]
labels = ['linUCB', 'random', 'greedy']
title = r'Metric: Cumulative regret'
ylabel = r'Cumulative regret'
xlabel = 'Rounds'
linUCB.plot_standard(data_plots, labels, [0], title, ylabel, xlabel, ax)
plt.show()
```
